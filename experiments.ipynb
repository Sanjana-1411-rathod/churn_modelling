Column Descriptions

RowNumber:The index or serial number of the row (not meaningful for analysis).

CustomerId:A unique identifier for each customer.

Surname:The last name of the customer (not useful for predictive modeling).

CreditScore:A numerical score representing the creditworthiness of the customer (higher is better).

Geography:The country where the customer is based (e.g., France, Spain).

Gender:	The gender of the customer (Male/Female).

Age:	The age of the customer in years.

Tenure:	The number of years the customer has been with the bank.

Balance	:The customer's account balance in the bank.

NumOfProducts:	The number of financial products the customer has with the bank.

HasCrCard:	Whether the customer has a credit card (1 = Yes, 0 = No).

IsActiveMember:	Whether the customer is an active member (1 = Yes, 0 = No).

EstimatedSalary:	The estimated annual salary of the customer.

Exited:	The target variable (1 = Customer left the bank, 0 = Customer stayed).

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data
```

```python
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
from sklearn.preprocessing import OneHotEncoder
Onehot_encoder_geo=OneHotEncoder()
geo_encoder=Onehot_encoder_geo.fit_transform(data[['Geography']]).toarray()
geo_encoder
```

```python
Onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns=Onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
data.head()
```

```python
with open('label_encoder_gender.pkl','wb')as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb')as file:
    pickle.dump(Onehot_encoder_geo,file)
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python

with open('scaler.pkl','wb')as file:
    pickle.dump(scaler,file)
    
```

ANN implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
(x_train.shape[1])
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[early_stopping_callback]
)
```

```python
model.save('model.h5')
```
